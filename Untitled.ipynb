Bollinger Bands

```python
df3.drop(columns=['Open', 'High', 'Low', 'Volume'], inplace=True)
```

```python
df3['MA20'] = df3['Close'].rolling(window=20).mean()
df3['20dSTD'] = df3['Close'].rolling(window=20).std() 

df3['upper'] = df3['MA20'] + (df3['20dSTD'] * 2)
df3['lower'] = df3['MA20'] - (df3['20dSTD'] * 2)
```

```python
df3['prev Close'] = df3['Close'].shift()
df3['prev upper'] = df3['upper'].shift()
df3['prev lower'] = df3['lower'].shift()
```

```python
df3['signal'] = np.where((df3['Close']> df3['upper'])&(df3['prev Close']< df3['prev upper']) ,-1,0)
df3['signal'] = np.where((df3['Close']< df3['lower'])&(df3['prev Close']> df3['prev lower']) ,1,df3['signal'])
```

```python
df3['signal'].value_counts()
```

```python
df3['position'] = df3['signal'].replace(to_replace=0,method='ffill')
df3['position'].value_counts()
```

```python
df3[['Close','MA20','upper','lower']].plot(figsize=(10,4))
plt.grid(True)
plt.title(' Bollinger Bands')
```

```python
df3['B&H_returns'] =df3['Close'].pct_change()
df3['strategy_returns'] = df3['B&H_returns']*df3['position'].shift(1)
```

```python
df3['B&H_cumulative'] = (1+df3['B&H_returns']).cumprod() -1
df3['strategy_cumulative'] = (1+df3['strategy_returns']).cumprod() -1
```

```python
df3[['B&H_cumulative','strategy_cumulative']].plot(figsize=(10,5))
```
