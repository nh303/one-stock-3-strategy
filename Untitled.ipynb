MACD

```python
df1=df.copy()
```

```python
df1[['Open', 'High', 'Low','Close']].plot(grid=True, figsize=(22,8))

```

3.Data Cleaning: 

```python
df1.drop(columns=['Open', 'High', 'Low', 'Volume'], inplace=True)
```

```python
df1['ema_26']=df1['Close'].ewm(span=26, adjust=False).mean()
df1['ema_12']=df1['Close'].ewm(span=12, adjust=False).mean()
df1['MACD']=df1['ema_12']-df1['ema_26']
                
df1['signal']=df1['MACD'].ewm(span=9, adjust=False).mean()

df1[['signal', 'MACD', 'Close']].plot(figsize=(12,9), grid=True, secondary_y='Close')

df1['position']=np.where(df1['MACD']>df1['signal'], 1, -1)
df1['position']=df1['position'].shift(1)

df1['cc_returns']=np.log(df1['Close']/df1['Close'].shift(1))
df1['strategy_returns']=df1['cc_returns']*df1['position']

df1['cumulative_returns']=(1+df1['strategy_returns']).cumprod()-1
df1['cumulative_returns'].plot(figsize=(12,8), grid=True)

df1[['cumulative_returns', 'position']].plot(figsize=(12,8), secondary_y='position', grid=True)
```

```python
df1['position'].value_counts()
```

4.Exploratory Data Analysis:

```python
import pyfolio as pf
a=pf.create_simple_tear_sheet(df1['cumulative_returns'], benchmark_rets=None)

```

EMA

```python
df2=df.copy()
```

3.Data Cleaning: 

```python
df2.drop(columns=['Open', 'High', 'Low', 'Volume'], inplace=True)
```

```python
df2['cc_returns']=np.log(df2['Close']/df2['Close'].shift(1))

df2['ema']=df2['Close'].ewm(span=20, adjust=False).mean()

df2.head()
```

```python
df2['position']=np.where((df2['Close']>df2['ema']), 1, 0)

print(df2)
```

```python
df2['position']=df2['position'].shift(1)

df2['position'].value_counts()
df2.dropna(inplace=True)
df2.head()
```

```python
df2['strategy_returns']=df2['cc_returns']*df2['position']

df2['strategy_returns']=1+df2['strategy_returns']
df2['cc_returns']=1+df2['cc_returns']

df2.head()
```

```python
df2['position'].value_counts()
```

```python
df2[['cc_returns', 'strategy_returns']].cumprod().plot(grid=True, figsize=(18, 9))

print('Buy and hold returns: ', np.round(df2['cc_returns'].cumprod()[-1], 2))
print('Strategy returns: ', np.round(df2['strategy_returns'].cumprod()[-1], 2))
```

BOLLINGER_BANDS

```python
df3=df.copy()
```

3.Data Cleaning: 

```python
df3.drop(columns=['Open', 'High', 'Low', 'Volume'], inplace=True)
```

```python
df3['MA20'] = df3['Close'].rolling(window=20).mean()
df3['20dSTD'] = df3['Close'].rolling(window=20).std() 

df3['upper'] = df3['MA20'] + (df3['20dSTD'] * 2)
df3['lower'] = df3['MA20'] - (df3['20dSTD'] * 2)
```

```python
df3['prev Close'] = df3['Close'].shift()
df3['prev upper'] = df3['upper'].shift()
df3['prev lower'] = df3['lower'].shift()
```

```python
df3['signal'] = np.where((df3['Close']> df3['upper'])&(df3['prev Close']< df3['prev upper']) ,-1,0)
df3['signal'] = np.where((df3['Close']< df3['lower'])&(df3['prev Close']> df3['prev lower']) ,1,df3['signal'])
```

```python
df3['signal'].value_counts()
```

```python
df3['position'] = df3['signal'].replace(to_replace=0,method='ffill')
df3['position'].value_counts()
```

```python
df3[['Close','MA20','upper','lower']].plot(figsize=(10,4))
plt.grid(True)
plt.title(' Bollinger Bands')
```

```python
df3['B&H_returns'] =df3['Close'].pct_change()
df3['strategy_returns'] = df3['B&H_returns']*df3['position'].shift(1)
```

```python
df3['B&H_cumulative'] = (1+df3['B&H_returns']).cumprod() -1
df3['strategy_cumulative'] = (1+df3['strategy_returns']).cumprod() -1
```

```python
df3[['B&H_cumulative','strategy_cumulative']].plot(figsize=(10,5))
```

4.Exploratory Data Analysis:

```python
import pyfolio as pf
c=pf.create_simple_tear_sheet(df3['strategy_returns'], benchmark_rets=None)
```
